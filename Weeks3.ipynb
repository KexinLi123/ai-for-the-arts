````python
# Define a function that returns a greeting
def say_hello(name="World"):
    """Return a greeting for the given name."""
    return f"Hello, {name}!"

# Use the function with the default parameter
print(say_hello())

# Use the function with a specific name
print(say_hello("Alice"))
````

### Step-by-step explanation of the function example

1. **Function definition**
   - `def say_hello(name="World"):` defines a function named `say_hello` with one parameter `name`.
   - The `name` parameter has a default value of `"World"`, so calling `say_hello()` without arguments uses that default.

2. **Docstring**
   - `"""Return a greeting for the given name."""` is a short description (optional) explaining what the function does.

3. **Return value**
   - `return f"Hello, {name}!"` builds and returns a string. `return` sends the result back to the caller; it does not print it.

4. **Calling the function**
   - `say_hello()` calls the function and returns the greeting for the default name.
   - `say_hello("Alice")` passes a specific argument and returns a personalized greeting.

5. **Why use functions?**
   - Functions group related code into a reusable block.
   - They make programs easier to read, test, and maintain.
   - Parameters let functions be flexible; return values let callers use results.

**Key aspects to notice:**
- A function does not run until you call it.
- Use parameters to pass data into functions.
- Use `return` to send results back; use `print()` when you want to display output.

## 4. While Loops
### What is a While Loop?
A while loop repeats code as long as a condition is true. Unlike a for loop, which runs a specific number of times, a while loop continues until you tell it to stop.

````python
# Example of a while loop
count = 0
while count < 5:
    print("Hello, world!")
    count = count + 1

print("\nLoop finished!")
````

### Step-by-step explanation of the while loop example

1. **Initialize a counter variable**
   - `count = 0` creates a variable that tracks how many times the loop has run.

2. **The while condition**
   - `while count < 5:` checks if `count` is less than 5.
   - If the condition is true, the loop runs; if false, the loop stops.

3. **Inside the loop**
   - `print("Hello, world!")` prints the message.
   - `count = count + 1` increases the counter by 1 each time (this is called incrementing).

4. **What happens**
   - First run: count = 0, print message, count becomes 1
   - Second run: count = 1, print message, count becomes 2
   - And so on... until count = 5
   - When count = 5, the condition `count < 5` is false, so the loop stops.

**Key aspects to notice:**
- You must manually update the counter variable inside the loop (like `count = count + 1`).
- If you forget to update the counter, the loop will run forever (infinite loop)!
- While loops are useful when you don't know in advance how many times you need to repeat.

---

## Comparison: For Loop vs While Loop

| Aspect | For Loop | While Loop |
|--------|----------|-----------|
| **When to use** | When you know how many times to repeat | When you want to repeat until a condition is met |
| **Example usage** | Repeat 5 times | Repeat until user enters "quit" |
| **How it counts** | Automatically manages a counter with `range()` | You must manually manage and update a counter |
| **Best for** | Iterating through a fixed number of items | Repeating based on a condition |
| **Code simplicity** | Simpler, cleaner code | Requires more manual setup |

**Example comparison:**

```python
# FOR LOOP - When you know how many times
for i in range(5):
    print("Hello, world!")

# WHILE LOOP - When you check a condition
count = 0
while count < 5:
    print("Hello, world!")
    count = count + 1
```

Both produce the same output, but use different approaches. Choose the one that makes most sense for what you're trying to do!
