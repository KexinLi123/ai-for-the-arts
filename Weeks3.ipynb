## Loops
### What is a While Loop?
A while loop repeats code as long as a condition is true. Unlike a for loop, which runs a specific number of times, a while loop continues until you tell it to stop.

````python
# Example of a while loop
count = 0
while count < 5:
    print("Hello, world!")
    count = count + 1

print("\nLoop finished!")
````

### Step-by-step explanation of the while loop example

1. **Initialize a counter variable**
   - `count = 0` creates a variable that tracks how many times the loop has run.

2. **The while condition**
   - `while count < 5:` checks if `count` is less than 5.
   - If the condition is true, the loop runs; if false, the loop stops.

3. **Inside the loop**
   - `print("Hello, world!")` prints the message.
   - `count = count + 1` increases the counter by 1 each time (this is called incrementing).

4. **What happens**
   - First run: count = 0, print message, count becomes 1
   - Second run: count = 1, print message, count becomes 2
   - And so on... until count = 5
   - When count = 5, the condition `count < 5` is false, so the loop stops.

**Key aspects to notice:**
- You must manually update the counter variable inside the loop (like `count = count + 1`).
- If you forget to update the counter, the loop will run forever (infinite loop)!
- While loops are useful when you don't know in advance how many times you need to repeat.

## Comparison: For Loop vs While Loop

| Aspect | For Loop | While Loop |
|--------|----------|-----------|
| **When to use** | When you know how many times to repeat | When you want to repeat until a condition is met |
| **Example usage** | Repeat 5 times | Repeat until user enters "quit" |
| **How it counts** | Automatically manages a counter with `range()` | You must manually manage and update a counter |
| **Best for** | Iterating through a fixed number of items | Repeating based on a condition |
| **Code simplicity** | Simpler, cleaner code | Requires more manual setup |

**Example comparison:**

```python
# FOR LOOP - When you know how many times
for i in range(5):
    print("Hello, world!")

# WHILE LOOP - When you check a condition
count = 0
while count < 5:
    print("Hello, world!")
    count = count + 1
```

Both produce the same output, but use different approaches. Choose the one that makes most sense for what you're trying to do!

````python
# Example: import and use the built-in math library

import math

# Use a function from the math library
pi = math.pi
print(f"Value of pi: {pi}")

# Use math functions
radius = 5
area = math.pi * radius ** 2
print(f"Area of circle with radius {radius}: {area}")

# Use more math functions
print(f"Square root of 16: {math.sqrt(16)}")
print(f"Absolute value of -7: {math.fabs(-7)}")

# You can also import specific items from a library
from math import sqrt, ceil

result = sqrt(25)
print(f"Square root of 25: {result}")

rounded_up = ceil(4.3)
print(f"Ceiling of 4.3: {rounded_up}")
````

### Step-by-step explanation of the import example

1. **Import a library**
   - `import math` — brings in the `math` library, which has math functions and constants built-in.
   - Without importing, you cannot use functions like `sqrt()` or constants like `pi`.

2. **Access library items using dot notation**
   - `math.pi` — accesses the constant `pi` from the `math` library. The dot (`.`) means "inside" or "from".
   - `math.sqrt(16)` — calls the `sqrt()` function from the `math` library to calculate square root.

3. **Using library functions for calculations**
   - `math.pi * radius ** 2` — uses the library's `pi` constant to calculate the area of a circle.
   - `math.sqrt()`, `math.fabs()`, etc. — are ready-to-use functions that do specific jobs.

4. **Import specific items directly**
   - `from math import sqrt, ceil` — instead of using `math.sqrt()`, you can now call `sqrt()` directly.
   - This is useful when you use the same function many times (saves typing `math.` each time).

5. **Why use libraries?**
   - Libraries are pre-written code that solve common problems.
   - They save you time — no need to write `sqrt()` yourself.
   - They are tested and reliable.
   - Python comes with many built-in libraries (like `math`, `random`, `datetime`).
   - You can also install and use external libraries (like `numpy`, `pandas`).

**Key aspects to notice:**
- `import` brings in a whole library; you must use dot notation (e.g., `math.sqrt`).
- `from ... import ...` brings in specific items; you can use them directly (e.g., `sqrt`).
- Libraries extend what Python can do without writing everything from scratch.
- Common built-in libraries: `math`, `random`, `datetime`, `os`, `sys`.

### Simple conditional example

This example shows how to use `if`, `elif`, and `else` to make decisions in your code — to run different code based on whether something is true or false. Run the code cell below to see how these statements work together.

````python
# Example 1: greeting based on time of day
time = 14  # represents hour in 24-hour format (14 = 2:00 PM)

if time < 12:
    print("Good morning!")
elif time < 18:
    print("Good afternoon!")
else:
    print("Good evening!")

print()  # blank line for readability

# Example 2: check a user's age and print a message
age = 25

if age < 13:
    print(f"You are {age} years old. You are a child.")
elif age < 18:
    print(f"You are {age} years old. You are a teenager.")
elif age < 65:
    print(f"You are {age} years old. You are an adult.")
else:
    print(f"You are {age} years old. You are a senior.")

print()  # blank line for readability

# Example 3: check score and assign a grade
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"Score: {score} -> Grade: {grade}")

print()  # blank line for readability

# Example 4: check if number is positive, negative, or zero
num = -5

if num > 0:
    print(f"{num} is positive")
elif num < 0:
    print(f"{num} is negative")
else:
    print(f"{num} is zero")
````

### Step-by-step explanation of if/elif/else

1. **The `if` statement — checks a single condition**
   - `if time < 12:` checks if the variable `time` is less than 12.
   - If true, the code inside the `if` block runs (print "Good morning!").
   - If false, Python skips to the next condition.

2. **The `elif` statement — checks additional conditions**
   - `elif time < 18:` means "else if" — only checked if the previous `if` was false.
   - If this condition is true, the code runs (print "Good afternoon!").
   - You can have multiple `elif` statements in a row.

3. **The `else` statement — fallback when all conditions are false**
   - `else:` runs if all previous `if` and `elif` conditions were false.
   - No condition is needed for `else` — it's the default/catch-all.
   - In the example, if time is 18 or later, "Good evening!" prints.

4. **How it flows**
   - Python evaluates conditions from top to bottom.
   - Once one condition is true and its code runs, Python skips all remaining `elif` and `else` blocks.
   - You only get one output per `if/elif/else` chain.

5. **Comparison operators used in conditions**
   - `<` less than
   - `>` greater than
   - `<=` less than or equal to
   - `>=` greater than or equal to
   - `==` equal to (note: two equals signs)
   - `!=` not equal to

6. **Indentation is critical**
   - Code inside `if`, `elif`, and `else` blocks must be indented (pushed to the right).
   - Python uses indentation to know which code belongs to which condition.

**Key aspects to notice:**
- `if` starts the decision chain.
- `elif` allows multiple conditions to be checked in order.
- `else` is optional and catches everything not caught by `if`/`elif`.
- Only one block runs per chain (the first true condition).
- Conditions evaluate to True or False.

## Operators

Operators are symbols that perform actions on variables and values. Python has several types: arithmetic (math), comparison (true/false checks), logical (and/or/not), and string operations (concatenation and repetition). Run the code cell below to see operators in action.

````python
# Example: Demonstrating different types of operators

print("=== 1. ARITHMETIC OPERATORS ===")
a = 10
b = 3

print(f"{a} + {b} = {a + b}")       # addition
print(f"{a} - {b} = {a - b}")       # subtraction
print(f"{a} * {b} = {a * b}")       # multiplication
print(f"{a} / {b} = {a / b}")       # division (returns float)
print(f"{a} // {b} = {a // b}")     # floor division (returns int)
print(f"{a} % {b} = {a % b}")       # modulo (remainder)
print(f"{a} ** {b} = {a ** b}")     # exponentiation (power)

print("\n=== 2. STRING OPERATIONS ===")
greeting = "Hello"
name = "World"

# String concatenation (addition)
full_greeting = greeting + ", " + name + "!"
print(f"Concatenation: {full_greeting}")

# String repetition (multiplication)
repeated = "Ha" * 3
print(f"Repetition: {repeated}")

# String with numbers (repetition)
line = "-" * 20
print(f"Line: {line}")

# You cannot add a string and number directly (will cause error)
try:
    result = "Hello" + 5  # This will fail
except TypeError as e:
    print(f"Error when adding string + number: {e}")

# But you can convert and then add
message = "I have " + str(5) + " apples"
print(f"Converted: {message}")

print("\n=== 3. COMPARISON OPERATORS ===")
x = 15
y = 10

print(f"{x} > {y}:  {x > y}")       # greater than
print(f"{x} < {y}:  {x < y}")       # less than
print(f"{x} >= {y}: {x >= y}")      # greater than or equal to
print(f"{x} <= {y}: {x <= y}")      # less than or equal to
print(f"{x} == {y}: {x == y}")      # equal to
print(f"{x} != {y}: {x != y}")      # not equal to

print("\n=== 4. LOGICAL OPERATORS ===")
age = 25
score = 85

# Logical AND (both must be true)
if age >= 18 and score >= 80:
    print(f"Age {age} and Score {score}: Both conditions are TRUE (and)")
else:
    print("Not both conditions")

# Logical OR (at least one must be true)
if age < 18 or score >= 80:
    print(f"Age {age} or Score {score}: At least one is TRUE (or)")

# Logical NOT (reverses true/false)
is_student = True
print(f"is_student: {is_student}")
print(f"not is_student: {not is_student}")

print("\n=== 5. OPERATOR PRECEDENCE ===")
result = 2 + 3 * 4
print(f"2 + 3 * 4 = {result}")  # Multiplication happens first (3*4=12), then 2+12=14

result = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result}")  # Parentheses first (2+3=5), then 5*4=20
````

### Step-by-step explanation of operators

**1. Arithmetic Operators (work with numbers)**

- `+` addition: `10 + 3 = 13`
- `-` subtraction: `10 - 3 = 7`
- `*` multiplication: `10 * 3 = 30`
- `/` division: `10 / 3 = 3.333...` (always returns float)
- `//` floor division: `10 // 3 = 3` (returns integer, rounds down)
- `%` modulo: `10 % 3 = 1` (remainder after division)
- `**` exponentiation: `10 ** 3 = 1000` (10 to the power of 3)

**2. String Operations (work with text)**

- **Concatenation (addition)**: `"Hello" + " " + "World"` produces `"Hello World"`
  - You combine strings together using the `+` operator.
  
- **Repetition (multiplication)**: `"Ha" * 3` produces `"HaHaHa"`
  - You repeat a string a certain number of times using `*` with a number.
  - Example: `"-" * 20` creates a line of 20 dashes.

- **Important**: You **cannot directly add a string and a number**
  - `"Hello" + 5` causes an error because Python doesn't know how to add text and a number.
  - **Solution**: Convert the number to a string first: `"Hello" + str(5)` produces `"Hello5"`

**3. Comparison Operators (return True or False)**

- `>` greater than: `15 > 10` is `True`
- `<` less than: `15 < 10` is `False`
- `>=` greater than or equal: `15 >= 15` is `True`
- `<=` less than or equal: `15 <= 20` is `True`
- `==` equal to: `15 == 15` is `True` (note: two equals signs)
- `!=` not equal to: `15 != 10` is `True`

**4. Logical Operators (combine True/False values)**

- `and`: Both conditions must be true. Example: `age >= 18 and score >= 80` is true only if both are true.
- `or`: At least one condition must be true. Example: `age < 18 or score >= 80` is true if either is true.
- `not`: Reverses the value. Example: `not True` is `False`, and `not False` is `True`.

**5. Operator Precedence (order of operations)**

Python follows standard math rules:
1. `**` (exponentiation) — highest priority
2. `*`, `/`, `//`, `%` (multiplication and division) — next
3. `+`, `-` (addition and subtraction) — lowest priority

Use parentheses `()` to override precedence: `(2 + 3) * 4` gives `20`, not `14`.

**Key aspects to notice:**
- Different operators work with different data types (numbers vs strings).
- String concatenation uses `+`, string repetition uses `*` with a number.
- Comparison operators return True or False — used in `if` statements.
- Logical operators (`and`, `or`, `not`) combine conditions for complex decisions.
- Always convert numbers to strings before concatenating with text.

## Classes and Objects

A class is a blueprint for creating objects. Think of a class like a cookie cutter — the cutter is the class, and each cookie made from it is an object. Classes let you bundle data (attributes) and functions (methods) together in a reusable way. Run the code cell below to see classes in action.

````python
# Example 1: Simple Greeter class
class Greeter:
    def __init__(self, name):
        """Initialize the Greeter with a name"""
        self.name = name
    
    def say_hello(self):
        """Return a greeting"""
        return f"Hello, {self.name}!"
    
    def say_goodbye(self):
        """Return a goodbye message"""
        return f"Goodbye, {self.name}!"

# Create instances (objects) from the Greeter class
greeter1 = Greeter("Alice")
greeter2 = Greeter("Bob")

print(greeter1.say_hello())    # Output: Hello, Alice!
print(greeter2.say_hello())    # Output: Hello, Bob!
print(greeter1.say_goodbye())  # Output: Goodbye, Alice!
````

````python
# Example 2: More advanced Person class
class Person:
    def __init__(self, name, age):
        """Initialize Person with name and age"""
        self.name = name
        self.age = age
    
    def introduce(self):
        """Return an introduction"""
        return f"My name is {self.name} and I am {self.age} years old."
    
    def have_birthday(self):
        """Increment age and return a message"""
        self.age += 1
        return f"{self.name} is now {self.age} years old."
    
    def change_name(self, new_name):
        """Change the person's name"""
        old_name = self.name
        self.name = new_name
        return f"{old_name} is now called {self.name}."

# Create a Person object
person1 = Person("Charlie", 25)
print(person1.introduce())         # Output: My name is Charlie and I am 25 years old.
print(person1.have_birthday())     # Output: Charlie is now 26 years old.
print(person1.change_name("Chuck")) # Output: Charlie is now called Chuck.
````

### Implementation 2: SECURITY

**Principle**: Validate input, handle errors gracefully, and protect against misuse.

**Goal**: Write code that fails safely and prevents common security issues.

**Key aspects**:

1. **Input Validation** — Check that input is the correct type (string) before using it.
   - `isinstance(message, str)` verifies the input is actually a string.
   - Prevents accidental misuse or intentional attacks using wrong types.

2. **Error Handling with Try-Except** — Catch potential errors and handle them gracefully.
   - If someone passes a number instead of text, the error is caught and reported clearly.
   - The program doesn't crash; it tells the user what went wrong.

3. **Type Hints and Docstrings** — Document what types are expected.
   - `message (str):` tells developers what type to use.
   - Docstrings explain what the function does and what can go wrong.

4. **Defensive Programming** — Assume something could go wrong and plan for it.
   - Check for empty strings: `if not message.strip():` prevents printing nothing.
   - Use descriptive error messages: `f"Expected string, but got {type(message).__name__}"` helps debugging.

5. **PEP 8 Compliance** — Follow Python's style guide for consistency and security.
   - Proper function naming, documentation, and error messages.

**Why this matters**:
- **Robustness** — Code handles edge cases gracefully.
- **Debugging** — Clear error messages help find and fix problems.
- **User protection** — Validates data before processing it.
- **Professional code** — Security-conscious code is expected in real-world applications.

---

````python
# Implementation 2: SECURITY
# Principle: Validate input, handle errors gracefully, and protect against misuse
# Goal: Write code that fails safely and prevents common security issues

def print_greeting(message):
    """
    Safely print a greeting message.
    
    Args:
        message (str): The message to print. Must be a string.
    
    Returns:
        None
        
    Raises:
        TypeError: If message is not a string.
    """
    # Validate input type
    if not isinstance(message, str):
        raise TypeError(f"Expected string, but got {type(message).__name__}")
    
    # Validate message is not empty
    if not message.strip():
        raise ValueError("Message cannot be empty or whitespace-only")
    
    # Safely escape and print
    print(message)

# Use the function with proper error handling
try:
    print_greeting("Hello, World!")
except (TypeError, ValueError) as e:
    print(f"Error: {e}")

# Demonstrate error handling
try:
    print_greeting(123)  # This will raise a TypeError
except TypeError as e:
    print(f"Caught error: {e}")
````
