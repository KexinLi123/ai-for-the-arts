## Text Processing Example: Word Frequency Counting

```python
from collections import Counter
import re
def process_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read().lower() # Remove punctuation and split into words
        words = re.findall(r'\w+', text)
        word_counts = Counter(words)
        for word, count in word_counts.most_common(10):
            print(f"{word}: {count}")
process_text ("some_text.txt")
# take page 1 of the labbook as example.
```

```python
## Image Processing Example: Grayscale Conversion and Basic Feature Extraction
```

```python
from PIL import Image
import numpy as np
def process_image(file_path):
    img = Image.open(file_path).convert("L") # convert to grayscale
    img_array = np.array(img)
    avg_intensity = np.mean(img_array)
    print(f"Average intensity: {avg_intensity}")
    img.save("processed_image.jpg")
process_image("sample_image.jpg")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

# Display original and processed images side-by-side (one row, two columns).
# This cell preserves your original `process_image` function above.

def show_before_after(original='sample_image.jpg', processed='processed_image.jpg'):
    if not os.path.exists(original):
        print(f"Original image not found: {original}")
        return
    if not os.path.exists(processed):
        print(f"Processed image not found: {processed}. Run process_image first.")
        return
    orig = Image.open(original)
    proc = Image.open(processed)
    orig_arr = np.array(orig)
    proc_arr = np.array(proc)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(orig_arr)
    axes[0].set_title('Original')
    axes[0].axis('off')
    # If processed image is grayscale (2D array), use gray colormap
    if proc_arr.ndim == 2:
        axes[1].imshow(proc_arr, cmap='gray')
    else:
        axes[1].imshow(proc_arr)
    axes[1].set_title('Processed')
    axes[1].axis('off')
    plt.tight_layout()
    plt.show()

# Call display function (adjust filenames if needed)
show_before_after()
```

```python
import librosa 
import numpy as np
def process_audio(file_path): # Load audio at a default sample rate 
    y, sr = librosa.load(file_path, sr=None)# Compute average amplitude (a rough measure of volume)
    avg_amplitude = np.mean(np.abs(y))
    print(f"Average amplitude: {avg_amplitude}")

    # Compute a short-time Fourier transform (STFT) for frequency analysis 
    D = np.abs(librosa.stft(y))
    freq_bins = np.mean(D, axis=1)
    dominant_frequency_bin = np.argmax(freq_bins)
    print(f"Dominant frequency bin: {dominant_frequency_bin}")

process_audio("audio1.ogg")
```
