## Text Processing Example: Word Frequency Counting

```python
from collections import Counter
import re
def process_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read().lower() # Remove punctuation and split into words
        words = re.findall(r'\w+', text)
        word_counts = Counter(words)
        for word, count in word_counts.most_common(10):
            print(f"{word}: {count}")
process_text ("some_text.txt")
# take page 1 of the labbook as example.
```

```python
from PIL import Image
import numpy as np
def process_image(file_path):
    img = Image.open(file_path).convert("L") # convert to grayscale
    img_array = np.array(img)
    avg_intensity = np.mean(img_array)
    print(f"Average intensity: {avg_intensity}")
    img.save("processed_image.jpeg")
process_image("sample_image.jpeg")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# This cell only displays the original and processed images side-by-side
# It preserves your original processing function (which saves processed_image.jpeg)

def show_before_after(original='sample_image.jpeg', processed='processed_image.jpeg'):
    orig = Image.open(original)
    proc = Image.open(processed)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(np.array(orig))
    axes[0].set_title('Original')
    axes[0].axis('off')
    proc_arr = np.array(proc)
    if proc_arr.ndim == 2:
        axes[1].imshow(proc_arr, cmap='gray')
    else:
        axes[1].imshow(proc_arr)
    axes[1].set_title('Processed')
    axes[1].axis('off')
    plt.tight_layout()
    plt.show()

# Run the display function (adjust paths if needed)
show_before_after()
```
