## Hello World

### Step-by-Step Code Breakdown

Let's break down the `print("Hello, World!")` line of code:

**1. `print` - The Print Function**
- `print` is a built-in function in Python
- A function is pre-written code that performs a specific task
- `print()` outputs information to the screen

**2. Parentheses `()`**
- Parentheses hold the information we want to pass to the function (called "arguments")
- They tell Python where the function begins and ends

**3. Quotation Marks `""`**
- Double quotes mark the beginning and end of a text string
- Everything inside the quotes is treated as text content
- Python also accepts single quotes `''`

**4. String Content**
- `Hello, World!` is the actual text we want to print
- This is a classic introductory program, traditionally the first program when learning a new programming language

### Execution Flow

When Python executes this line of code:
1. Python recognizes the `print` function
2. It sees the parentheses, knowing there are arguments to pass
3. It extracts the text `Hello, World!` from the double quotes
4. It outputs this text to the screen/console

### Simple Variations

Try modifying the string content:

```python
# Example 1: Basic Hello World
print("Hello, World!")

# Example 2: Using single quotes
print('Hello, World!')

# Example 3: Printing different content
print("Welcome to learning Python!")

# Example 4: Printing multiple lines
print("First line")
print("Second line")
print("Third line")
```
