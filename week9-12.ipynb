# Designing a Creative Use of a Large Language Model

## 1. Introduction

This notebook presents the design of a creative application that uses a large language model (LLM). 
The application is designed to support learning and cultural engagement by allowing users to interact with historical perspectives through AI-generated dialogue.
This is a conceptual design only and does not include technical implementation.

## 2. What is a Large Language Model?

A large language model (LLM) is a type of artificial intelligence trained on very large collections of text such as books, articles, and websites. 
It learns patterns in language and generates responses by predicting the most likely next word in a sequence.
Although LLMs can produce human-like text, they do not have true understanding and may reflect biases present in their training data.


## 3. Description of the Proposed Application

The proposed application is an AI-powered educational tool that allows students to engage in simulated conversations with historical figures or voices from underrepresented communities.
The aim is to make history and cultural learning more engaging and accessible.


### Target Users

The primary users of this application are university students and secondary school students studying history, cultural studies, or social sciences.
It may also be used by museum visitors or members of the public interested in learning about the past in an interactive way.


### Human–AI Interaction

Users interact with the system by typing questions or prompts, such as asking about daily life, historical events, or personal experiences.
The large language model generates responses in the voice of a specific historical perspective, based on available historical records.
This conversational format encourages curiosity and deeper engagement with learning materials.


#### Example of Human–AI Interaction (Simulated)

**User:** What was daily life like for a dock worker in colonial-era London?

**AI (LLM):** Based on available historical records, dock work was physically demanding and often insecure. Employment depended on ship arrivals, and many workers faced irregular wages. It is important to note that these accounts are based on surviving sources and may not represent all experiences.


## 4. Role of the Large Language Model

In this application, the large language model (LLM) is used to generate conversational responses based on user input.
The model produces text by predicting likely word sequences from patterns learned during training, rather than by retrieving verified factual information.
As a result, the responses may sometimes be incomplete, biased, or factually inaccurate, even when they appear fluent and confident.

Instead of treating the LLM as an authoritative source, the design intentionally presents it as a tool for exploration and discussion.
The application can include clear disclaimers, prompts that encourage users to question the responses, and suggestions to compare AI-generated content with trusted academic or archival sources.
By making the limitations of the LLM visible, the design supports the development of critical thinking skills, encouraging students to evaluate information, identify uncertainty, and reflect on how knowledge is constructed rather than passively accepting AI outputs.




## 5. Learning, Cultural, and Inclusion Benefits

This design enhances learning by allowing users to actively explore historical topics through dialogue rather than passive reading.
It can increase cultural engagement by presenting perspectives that are often marginalised in traditional educational materials.
By including voices from underrepresented groups, the application supports inclusion and encourages users to reflect on whose stories are told in history.


## 6. Critical Reflection

There are important ethical and social considerations in using large language models in this way.
The training data of LLMs may reflect cultural biases, which could lead to inaccurate or stereotypical representations.
There are also concerns about data privacy when using cloud-based AI services, as it is often unclear how user data is stored or reused.
Additionally, allowing an AI to speak on behalf of historical or marginalised groups raises questions about authenticity and authority.

To clarify these concerns, the table below summarises key risks associated with the use of large language models and the corresponding design responses.

| Potential Risk | Design Response |
|---------------|----------------|
| Hallucinated or inaccurate content | Encourage users to verify information with trusted sources |
| Cultural or historical bias | Make limitations of training data explicit |
| Over-trust in AI-generated responses | Present AI output as exploratory rather than authoritative |


## 7. Conclusion

This notebook has presented a conceptual design for a creative application of large language models in education.
While the design offers opportunities for enhanced learning and inclusion, it also requires careful consideration of ethical and cultural implications.
Future work could explore how smaller or locally hosted models might address some of these concerns.
